Tujuh suara tersebut masing-masing di ekstraksi fiturnya. Untuk kasus ini saya menggunakan fitur Average Energy, pada library librosa namanya RMS(Root Mean Square) Energy



```python
cat1, sr= librosa.load('classifierData/duck/duck1.wav', duration=2.0) #load audio
cat1_train = librosa.feature.rmse(y=cat1) #ekstraksi fitur
#plot
S = librosa.magphase(librosa.stft(cat1, window=np.ones, center=False))[0]
rms = librosa.feature.rmse(S=S)
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogy(rms.T, label='Average Energy cat1.wav')
plt.xticks([])
plt.xlim([0, rms.shape[-1]])
plt.legend(loc='best')
plt.subplot(2, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time')
plt.title('log Power spectrogram')
plt.tight_layout()
```

### Training Data
Sebelum komputer bisa mengenali data maka kita latih komputer terlebih dahulu untuk bisa mengenali pola suatu data. Saya menggunakan MLPClassifier(), neural network classifier dari scikit-learn. Saya memilih neural network karena data yang akan ditrain tidak terlalu banyak. Pada proses training, komputer akan mengingat fitur dari ekstraksi data (train_x) yang akan menghasilkan label suara yang sesuai (train_y) agar bisa mengenali polanya kemudian.

```python
#menggabungkan data menjadi satu numpy array data train
train_x = [cat1_train.ravel(), cat2_train.ravel(), dog1_train.ravel(), duck1_train.ravel(), duck2_train.ravel(), horse1_train.ravel(), horse2_train.ravel()]
#melabeli fitur data
train_y = np.array(['cat','cat','dog','duck','duck','horse','horse'], dtype=object)
```

```python
mlp = MLPClassifier()
mlp.fit(train_x,train_y)
```
